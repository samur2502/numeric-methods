# Equações diferenciais

## Método de Euler

```python
def f(x,t):
    return -x**3 + np.sin(t)

def Euler(x, a, b, N, func):
    
    t = np.linspace(a,b,N)
    h = t[1]-t[0]

    X = np.zeros(N)
    for i in range(N):
        X[i] = x
        x += h*func(x,t[i])
    return X, t
```

```python
X, T = Euler(0, 0, 10, 10000, f)
plt.plot(T, X)
plt.show()
```

## Método de Runge-Kutta

### Para 1 derivada 1 grau (2ª ordem de Runge-Kutta)

```python
def f(x,t):
    return -x**3 + np.sin(t)

def Runge_Kutta(x, a, b, N, func):
    t = np.linspace(a, b, N)
    h = t[1]-t[0]
    X = np.zeros(N)
    for i in range (N):
        X[i] = x
        k1 = func(x,t[i]) * h
        k2 = func(x + 0.5 * k1, t[i] + h/2) * h
        x += k2
    return X, t
```

```python
X, T = Euler(0, 0, 10, 100, f)
plt.plot(T, X)
plt.show()
```

### Para 1 derivada 1 grau (4ª ordem de Runge-Kutta)


```python
def f(x,t):
    return -x**3 + np.sin(t)

def Runge_Kutta(x, a, b, N, func):
    t = np.linspace(a, b, N)
    h = t[1]-t[0]
    X = np.zeros(N)
    for i in range (N):
        X[i] = x
        k1 = f(x, t[i]) * h
        k2 = func(x + 0.5*k1, t[i] + h/2) * h 
        k3 = func(x + 0.5*k2, t[i] + h/2) * h
        k4 = func(x + k3, t[i] + h) * h
        x += (k1 + 2*k2 + 2*k3 + k4) / 6
    return X, t
```

```python
X, T = Runge_Kutta(0, 0, 10, 100, f)
plt.plot(T, X)
plt.show()
```

### Para 2 derivadas 1 grau (4ª ordem de Runge-Kutta)


```python
def f(r,t):
    alpha = 1
    betta = gamma = 0.5
    delta = 2
    x, y = r[0], r[1]
    dx = alpha*x - betta*x*y                # dx/dt
    dy = gamma*x*y - delta*y                # dy/dt
    return np.array([dx,dy],float)

def Runge_Kutta(x, y, a, b, N, func):
    r = np.array([x,y], float)
    t = np.linspace(a, b, N)
    h = t[1]-t[0]
    X = np.zeros(N)
    Y = np.zeros(N)
    
    for i in range (N):
        X[i], Y[i] = r[0], r[1]
        k1 = f(r, t[i]) * h
        k2 = f(r + 0.5*k1, t[i] + h/2) * h
        k3 = f(r + 0.5*k2, t[i] + h/2) * h
        k4 = f(r + k3, t[i] + h) * h
        r += (k1 + 2*k2 + 2*k3 + k4) / 6
    return X, Y, t
```

```python
X, Y, T = Runge_Kutta(2, 2, 0, 30, 10000, f)

plt.plot(T, X, label = "x(t)")
plt.plot(T, Y, label = "y(t)")
plt.legend()
plt.show()
```

### Para 3 derivadas 1 grau (4ª ordem de Runge-Kutta)


```python
def f(r,t):
    sigma = 10
    r1 = 28
    b = 8/3
    x, y, z = r[0], r[1], r[2]                               
    dx = sigma*(y-x)                       # dx/dt
    dy = r1*x-y-x*z                         # dy/dt
    dz = x*y-b*z                           # dz/dt
    return np.array([dx,dy,dz],float)

def Runge_Kutta(x, y, z, a, b, N, func):
    r = np.array([x,y,z], float)
    t = np.linspace(a, b, N)
    h = t[1]-t[0]
    X = np.zeros(N)
    Y = np.zeros(N)
    Z = np.zeros(N)
    
    for i in range (N):
        X[i], Y[i], Z[i] = r[0], r[1], r[2]
        k1 = f(r, t[i]) * h
        print(k1,"\n")
        k2 = f(r + 0.5*k1, t[i] + h/2) * h
        k3 = f(r + 0.5*k2, t[i] + h/2) * h
        k4 = f(r + k3, t[i] + h) * h
        r += (k1 + 2*k2 + 2*k3 + k4) / 6
    return X, Y, Z, t
```

```python
X, Y, Z, T = Runge_Kutta(0,1,0,0,50,10000, f)
plt.plot(T, Y, label = f"y(t)")
plt.legend()
plt.show()
```

### Para 1 derivadas 2 grau (4ª ordem de Runge-Kutta)

```python
def f(r,t):
    x, y = r[0], r[1]
    dx = y
    return dx

def Runge_Kutta(x, y, a, b, N, func):
    r = np.array([x,y], float)
    t = np.linspace(a, b, N)
    h = t[1]-t[0]
    X = np.zeros(N)
    Y = np.zeros(N)
    
    for i in range (N):
        X[i], Y[i] = r[0], r[1]
        k1 = f(r, t[i]) * h
        k2 = f(r + 0.5*k1, t[i] + h/2) * h
        k3 = f(r + 0.5*k2, t[i] + h/2) * h
        k4 = f(r + k3, t[i] + h) * h
        r += (k1 + 2*k2 + 2*k3 + k4) / 6
    return X, Y, t
```

```python
X, Y, T = Runge_Kutta(179*np.pi/180, 0.01, 0, 20, 10000, f)
plt.plot(T, X)
plt.show()
```
