## 
# <h1><center>Heat Equation</center></h1>

$$
\frac{\partial u \left (\textbf{r},\; t \right )} {\partial t} = \nabla \cdot  \left [ \text{D} \left ( u\left (\textbf{r},\; t \right ),\; r \right )\; \nabla u \left (\textbf{r},\; t \right ) \right ]
$$

where $u \left (\textbf{r},\; t \right )$ is the density of the diffusing material at location $\textbf{r}$ and time $t$ and $\text{D} \left ( u\left (\textbf{r},\; t \right ),\; r \right )$ is the collective diffusion coefficient for density $u$ at location $\textbf{r}$ and $\nabla$ represents the vector differential operator nabla. If the diffusion coefficient depends on the density then the equation is nonlinear, otherwise it is linear and we have the heat equation:

$$
\frac{\partial u \left (\textbf{r},\; t \right )} {\partial t} = D \cdot \nabla^{2} u \left (\textbf{r},\; t \right )
$$

## 
# <h3><center>Heat Equation in 1D</center></h3>

$$
\frac{\partial u \left (x,\; t \right )} {\partial t} = D \frac{\partial^2 u \left (x,\; t \right )} {\partial x^2}
$$
on the interval $x \in [0, L]$ with initial condition

$$
u(x, 0) = f (x),\: \forall  x \in [0, L]
$$

#### Dirichlet boundary conditions

$$u(0, t) = c_1, \: u(L, t) = c_2 \quad \forall \, t \geq0$$

#### Von-Neumann boundary conditions

$$\frac{\partial u}{\partial x}\bigg|_{x=0} = c_1 \quad \forall \, t \geq 0$$

$$\frac{\partial u}{\partial x}\bigg|_{x=L} = c_2 \quad \forall \, t \geq 0$$

#### Robin boundary conditions

$$\frac{\partial u}{\partial x}\bigg|_{x=0} + c_1 u(0, t) =  c_2\quad \forall \, t \geq 0$$

$$\frac{\partial u}{\partial x}\bigg|_{x=L} + c_3 u(L, t) =  c_4\quad \forall \, t \geq 0$$



#### Forward Euler

$$
\frac{u^{n+1}_i-u^n_i}{\Delta t} = D \frac{u^{n}_{i+1} - 2u^n_i + u^n_{i-1}}{\Delta x^2}
$$

$$
u^{n+1}_i = u^n_i + F\left(u^{n}_{i+1} - 2u^n_i + u^n_{i-1}\right) \:, \:\:\: F = D\frac{\Delta t}{\Delta x^2}
$$


For the Von-Neumann boundry condition at $x=0$:
$$\frac{\partial u}{\partial x}\bigg|_{x=0} \approx  \frac{u^n_{1} - u^n_{-1}}{2\Delta x} = c_1 \: \Leftrightarrow \: u^n_{-1} =u^n_{1} -  2 c_1 \Delta x  $$


Plugging it to the equation, it comes that:

$$
u^{n+1}_0 = u^n_0 + 2F\left(u^{n}_{1} - u^n_0  -  c_1 \Delta x\right)
$$

Similar to $x=L$:

$$\frac{\partial u}{\partial x}\bigg|_{x=L} \approx  \frac{u^n_{N_x+1} - u^n_{N_x-1}}{2\Delta x} = c_2 \: \Leftrightarrow \: u^n_{N_x+1} =u^n_{N_x-1} +  2 c_2 \Delta x  $$

$$
u^{n+1}_{N_x} = u^n_{N_x} + 2F\left(u^n_{N_x-1} - u^n_{N_x} + c_2 \Delta x\right) 
$$


Now, the Robin boundry condition at $x=0$:

$$\frac{\partial u}{\partial x}\bigg|_{x=0} \approx  \frac{u^n_{1} - u^n_{-1}}{2\Delta x} + c_1 u^n_{0} = c_2 \: \Leftrightarrow \: u^n_{-1} = u^n_{1} + 2c_1\Delta x u^n_{0} - 2 c_2 \Delta x$$


It comes that:
$$
u^{n+1}_0 = u^n_0 + 2F\left(u^{n}_{1} + u^n_{0} (c_1\Delta x - 1) - c_2 \Delta x\right)
$$
Similar to $x=L$:

$$\frac{\partial u}{\partial x}\bigg|_{x=L} \approx  \frac{u^n_{N_x+1} - u^n_{N_x-1}}{2\Delta x} + c_3 u^n_{N_x} = c_4 \: \Leftrightarrow \: u^n_{N_x+1} = u^n_{N_x-1} - 2c_1\Delta x u^n_{N_x} + 2 c_2 \Delta x$$

$$
u^{n+1}_{N_x} = u^n_{N_x} + 2F\left(u^{n}_{N_x-1} + u^n_{N_x} (-c_1\Delta x - 1) + c_2 \Delta x\right)
$$

#### Stability of explicit method

Since the solution is a function of time and space, we shall assume that the solution is subject to a perturbation given by $ u^k_i = A(t) \cos(\omega x_i)$, where $ A(t) $ is the time varying amplitude and $ \omega $ is the spatial frequency of the disturbance. We then have the following:

$$
u^k_{i-1} = A(t) \cos[\omega(x_i - \Delta x)] = A(t) [\cos \omega x_i \cos \omega \Delta x - \sin \omega x_i \sin \omega \Delta x],
$$

$$
u^k_{i+1} = A(t) \cos[\omega(x_i + \Delta x)] = A(t) [\cos \omega x_i \cos \omega \Delta x + \sin \omega x_i \sin \omega \Delta x],
$$

$$
u^{k+1}_i = A(t + \Delta t) \cos(\omega x_i).
$$

We substitute these, and after simplification, we get:

$$
\frac{A(t + \Delta t)}{A(t)} = 1 - 2F(1 - \cos \omega \Delta x)
$$

Note that the maximum value of $1 - \cos \omega \Delta x$ is 2, and the amplitude ratio $ |G| = \left| \frac{A(t + \Delta t)}{A(t)} \right| $ is less than unity (this is the necessary condition for stability) only if $ F < 0.5 $. Thus, the explicit method is conditionally stable.
Unfortunately the scheme is unstable unless $\Delta t$ is made so small that the simulation is impossibly slow.

#### Backward Euler

$$ 
\frac{u^{n}_i-u^{n-1}_i}{\Delta t} = D \frac{u^{n}_{i+1} - 2u^n_i + u^n_{i-1}}{\Delta x^2}
$$

$$  
- F u^n_{i-1} + \left(1+  2F \right) u^{n}_i - F u^n_{i+1} = u_{i}^{n-1}
$$

$$ AU = b, \:\: U=(u^n_1,\ldots,u^n_{N_x-1})$$

$$
A_{i,i-1} = -F, \quad A_{i,i} = 1 + 2F, \quad A_{i,i+1} = -F
$$
$$
b_i = u^{n-1}_i
$$

The Von-Neumann boundry condition from the previous result:
$$\frac{\partial u}{\partial x}\bigg|_{x=0} \approx  \frac{u^n_{1} - u^n_{-1}}{2 \Delta x} = -c_1 \: \Leftrightarrow \: u^n_{-1} = u^n_{1} +  2 c_1 \Delta x$$

$$
- 2F u^n_{1}  + \left(1+  2F \right) u^{n}_0 = u_{0}^{n-1} + 2 c_1 F \Delta x 
$$


$$\frac{\partial u}{\partial x}\bigg|_{x=L} \approx  \frac{u^n_{N_x+1} - u^n_{N_x-1}}{2\Delta x} = c_2 \: \Leftrightarrow \: u^n_{N_x+1} =u^n_{N_x-1} + 2c_2\Delta x$$

$$
- 2F u^n_{N_x-1} + \left(1+  2F \right) u^{n}_{N_x} = u_{N_x}^{n-1} + 2c_2F\Delta x 
$$


Now we consider the Robin boundry condition:
$$\frac{\partial u}{\partial x}\bigg|_{x=0} \approx  \frac{u^n_{1} - u^n_{-1}}{2\Delta x} = - c_1 u^n_{0} + c_2 \: \Leftrightarrow \: u^n_{-1} = u^n_{1} + 2 c_1 \Delta x u^n_{0} - 2 c_2 \Delta x$$

$$
- 2F u^n_{1}  + \left(1+  2F - 2c_1 F\Delta x \right) u^{n}_0 + 2 c_2 F \Delta x= u_{0}^{n-1}
$$

For $x=L$:

$$
- 2F u^n_{N_x-1} + \left(1+  2F + 2\kappa F\Delta x\right) u^n_{N_x} = u_{N_x}^{n-1}
$$


#### Stability of implicit method

The amplification factor $ |G| $ is given by:

$$
|G| = \left| \frac{A(t + \Delta t)}{A(t)} \right| = \left| \frac{1}{1 + 2F(1 - \cos \omega \Delta x)} \right|
$$

This is less than 1 for all values of $ F $, indicating that the implicit method is unconditionally stable.

##### Crank-Nicholson (CN)

$$\frac{\partial u}{\partial x} \approx \frac{u^{n+\frac{1}{2}}_{i+1} - u^{n+\frac{1}{2}}_{i-1}}{2\Delta x}$$

$$\frac{\partial^2 u}{\partial x^2} \approx \frac{1}{\Delta x^2}\left(u^{n+\frac{1}{2}}_{i-1} - 2u^{n+\frac{1}{2}}_i + u^{n+\frac{1}{2}}_{i+1}\right)$$

$$ u^{n+\frac{1}{2}}_i\approx \frac{1}{2}\left(u^{n}_i +u^{n+1}_{i}\right)$$

$$ 
\frac{u^{n+1}_i-u^{n}_i}{\Delta t} = D \frac{u^{n+1}_{i-1} - 2u^{n+1}_i + u^{n+1}_{i+1}}{\Delta x^2} 
$$

$$
u^{n+1}_i - \frac{1}{2} F(u^{n+1}_{i-1} - 2u^{n+1}_i + u^{n+1}_{i+1})
= u^{n}_i + \frac{1}{2} F(u^{n}_{i-1} - 2u^{n}_i + u^{n}_{i+1})
$$

$$
A_{i,i-1} = -\frac{1}{2} F
$$
$$
A_{i,i} = 1 + F
$$
$$
A_{i,i+1} = -\frac{1}{2} F
$$

$$
B_{i,i-1} = \frac{1}{2} F
$$
$$
B_{i,i} = 1 - F
$$
$$
B_{i,i+1} = \frac{1}{2} F
$$


The Von-Neumann boundry condition from the previous result:

$$\frac{\partial u}{\partial x}\bigg|_{x=0} \approx \frac{u^{n+\frac{1}{2}}_{1} - u^{n+\frac{1}{2}}_{-1}}{2\Delta x} = -q \: \Leftrightarrow \:  \frac{1}{2}\left(u^{n}_{-1} +u^{n+1}_{-1}\right) = \frac{1}{2}\left(u^{n}_1 +u^{n+1}_{1}\right)+2 q\Delta x $$

$$  u^{n+1}_{-1}= u^{n}_1 +u^{n+1}_{1}-u^{n}_{-1} +4 q\Delta x $$

$$
u^{n+1}_0 - \frac{1}{2} F(u^{n+1}_{-1} - 2u^{n+1}_0 + u^{n+1}_{1})
= u^{n}_0 + \frac{1}{2} F(u^{n}_{-1} - 2u^{n}_0 + u^{n}_{1})
$$

$$
u^{n+1}_0 + F(u^{n+1}_0 - u^{n+1}_{1})
= u^{n}_0 - F(u^{n}_0 - u^{n}_{1}) +2 F q\Delta x 
$$

Same can be done to $x=L$:

$$
u^{n+1}_{N_x}+ F(u^{n+1}_{N_x} - u^{n+1}_{{N_x}-1})
= u^{n}_{N_x} - F(u^{n}_{N_x} - u^{n}_{{N_x}-1}) +2 F q\Delta x 
$$

Now we consider the Robin boundry condition:

$$\frac{\partial u}{\partial x}\bigg|_{x=0} \approx \frac{u^{n+\frac{1}{2}}_{1} - u^{n+\frac{1}{2}}_{-1}}{2\Delta x} = \kappa u^n_{0} \: \Leftrightarrow \:  \frac{1}{2}\left(u^{n}_{-1} +u^{n+1}_{-1}\right) = \frac{1}{2}\left(u^{n}_1 +u^{n+1}_{1}\right) -  2 \kappa\Delta x u^n_{0}$$

$$  u^{n+1}_{-1}= u^{n}_1 +u^{n+1}_{1}-u^{n}_{-1} - 4 \kappa\Delta x u^n_{0}$$

$$
u^{n+1}_0 - \frac{1}{2} F(u^{n+1}_{-1} - 2u^{n+1}_0 + u^{n+1}_{1})
= u^{n}_0 + \frac{1}{2} F(u^{n}_{-1} - 2u^{n}_0 + u^{n}_{1})
$$

$$
u^{n+1}_0 + F(u^{n+1}_0 - u^{n+1}_{1})
= u^{n}_0 - F(u^{n}_0 - u^{n}_{1}) -2 \kappa F \Delta x u^n_{0} 
$$

Same can be done to $x=L$:

$$
u^{n+1}_{N_x}+ F(u^{n+1}_{N_x} - u^{n+1}_{{N_x}-1})
= u^{n}_{N_x} - F(u^{n}_{N_x} - u^{n}_{{N_x}-1}) -2 \kappa F \Delta x u^n_{N_x}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
plt.rcParams['animation.embed_limit'] = 1024

import scipy.sparse as sp
import scipy.sparse.linalg as spla

class DiffusionSolver:
    def __init__(self, L=1.0, T=5, D = 0.1, Nx=128, Nt=256, BC=('Robin')):
        
        # Parameters
        self.L = L  # Length of the domain
        self.T = T  # Total simulation time
        self.D = D  # Diffusion coefficient
        self.Nx = Nx  # Number of spatial points
        self.Nt = Nt  # Number of time steps

        # Initialize spatial and time points
        self.x = np.linspace(0, L, Nx)
        self.t = np.linspace(0, T, Nt)
        self.u = np.zeros((self.Nx, self.Nt))

        self.dx = np.diff(self.x)[0]  # Spatial step |size
        self.dt = np.diff(self.t)[0]  # Time step size

        self.BC = BC # Boundary conditions

        self.F = self.D * self.dt / self.dx**2

        # Initialize the quantity being diffused
        self.u[:,0] = self.initial_condition()
    
    def initial_condition(self):
        # Define the initial condition 
        #return np.exp(-50 * (self.x - self.L/2)**2) + 10 # Gaussian distribution
        #return np.ones(self.Nx)
        return np.sin(np.pi*self.x)

    def solve_FTCS(self, BC_values = (0,0,0,0)):
        # Solve the diffusion equation using the forward Euler method
        c1, c2, c3, c4 = BC_values

        if self.F>0.5:
            print(f'Unstable - F={self.F.round(3)}')

        if self.BC == 'Dirichlet':
            for n in range(self.Nt - 1):
                self.u[1:-1, n+1] = self.u[1:-1, n] + self.F*(self.u[2:, n] - 2*self.u[1:-1, n] + self.u[:-2, n])
                self.u[0,:] = c1
                self.u[-1,:] = c2
                
        elif self.BC == 'Von-Neumann':
            for n in range(self.Nt - 1):
                self.u[1:-1, n+1] = self.u[1:-1, n] + self.F*(self.u[2:, n] - 2*self.u[1:-1, n] + self.u[:-2, n])
                self.u[0,n+1] = self.u[0, n] + 2*self.F*(self.u[1, n] - self.u[0, n] - c1*self.dx)
                self.u[-1,n+1] = self.u[-1, n] + 2*self.F*(self.u[-2, n] - self.u[-1, n] + c2*self.dx)
                
        elif self.BC == 'Robin':
            for n in range(self.Nt - 1):
                self.u[1:-1, n+1] = self.u[1:-1, n] + self.F*(self.u[2:, n] - 2*self.u[1:-1, n] + self.u[:-2, n])
                self.u[0,n+1] = self.u[0, n] + 2*self.F*(self.u[1, n] + self.u[0, n] * (c1*self.dx - 1) - c2*self.dx)
                self.u[-1,n+1] = self.u[-1, n] + 2*self.F*(self.u[-2, n] + self.u[-1, n]* (c3*self.dx - 1) + c4*self.dx)
    
    def solve_BTCS(self, BC='Von-Neumann', BC_left = 0, BC_right = 0):
        # Solve the diffusion equation using the backward Euler method
        A = sp.diags([-self.F, 1+2*self.F, -self.F], [-1, 0, 1], shape=(self.Nx, self.Nx)).tolil()
        b = np.zeros((self.Nx,1))

        if BC == 'Dirichlet':
            A[0,0] = A[-1,-1] = 1
            A[0,1] = A[-1,-2] = 0
            A = A.tocsr()

            for n in range(self.Nt-1):
                b[:,0] = self.u[:,n]
                b[0,:] = BC_left
                b[-1,:] = BC_right
                self.u[:,n+1] = spla.spsolve(A,b)

        elif BC == 'Von-Neumann':
            A[0,0] = A[-1,-1] = 1 + 2*self.F
            A[0,1] = A[-1,-2] = -2 * self.F
            A = A.tocsr()
    
            for n in range(self.Nt-1):
                b[:,0] = self.u[:,n]
                b[0,:] += 2 * self.F * BC_left * self.dx
                b[-1,:] += 2 * self.F * BC_right * self.dx
                self.u[:,n+1] = spla.spsolve(A,b)

        elif BC == 'Robin':
            A[0,0] = 1 + 2*self.F + 2 * BC_left * self.F * self.dx
            A[-1,-1] = 1 + 2*self.F + 2 * BC_right * self.F * self.dx
            A[0,1] = A[-1,-2] = -2 * self.F
            A = A.tocsr()
    
            for n in range(self.Nt-1):
                b[:,0] = self.u[:,n]
                self.u[:,n+1] = spla.spsolve(A,b)
    
    def solve_CN(self, BC='Von-Neumann', BC_left = 0, BC_right = 0):
        # Solve the diffusion equation using the Crank-Nicholson method
        A = sp.diags([-self.F/2, 1+self.F, -self.F/2], [-1, 0, 1], shape=(self.Nx, self.Nx)).tolil()
        B = sp.diags([self.F/2, 1-self.F, self.F/2], [-1, 0, 1], shape=(self.Nx, self.Nx)).tolil()
        b = np.zeros((self.Nx,1))

        if BC == 'Dirichlet':
            A[0,0] = A[-1,-1] = 1
            A[0,1] = A[-1,-2] = 0
            A = A.tocsr()
            B = B.tocsr()

            for n in range(self.Nt-1):
                b[:,0] = B @ self.u[:,n]
                b[0,:] = BC_left
                b[-1,:] = BC_right
                self.u[:,n+1] = spla.spsolve(A,b)

        elif BC == 'Von-Neumann':
            A[0,0] = A[-1,-1] = 1 + self.F
            A[0,1] = A[-1,-2] = -self.F
            A = A.tocsr()
            B[0,0] = B[-1,-1] = 1 - self.F
            B[0,1] = B[-1,-2] = self.F
            B = B.tocsr()
            
            for n in range(self.Nt-1):
                b[:,0] = B @ self.u[:,n]
                b[0,:] += 2 * self.F * BC_left * self.dx
                b[-1,:] += 2 * self.F * BC_right * self.dx
                self.u[:,n+1] = spla.spsolve(A,b)
    
        elif BC == 'Robin':
            A[0,0] = A[-1,-1] = 1 + self.F
            A[0,1] = A[-1,-2] = -self.F
            A = A.tocsr()
            B[0,0] = 1 - self.F - 2 * BC_left * self.F * self.dx
            B[-1,-1] = 1 - self.F - 2 * BC_right * self.F * self.dx
            B[0,1] = B[-1,-2] = self.F

            B = B.tocsr()
            
            for n in range(self.Nt-1):
                b[:,0] = B @ self.u[:,n]
                self.u[:,n+1] = spla.spsolve(A,b)
                
    def plot_animation(self):
        fig, ax = plt.subplots(figsize=(15,8))
        
        ax.plot(self.x, self.u[:, 0], '-', color='black',label='Initial Condition')
        line, = ax.plot(self.x, self.u[:, 0], color='red', label='Diffusion')

        ax.set_xlabel('x')
        ax.set_ylabel('u')
        ax.legend()
        ax.grid()

        y_min = min(self.u.min()*1.1, 0)
        y_max = max(self.u.max()*1.1, 1)
        
        ax.set_ylim(y_min, y_max)
        
        # Define the update function for the animation
        def update(frame):
            line.set_data(self.x, self.u[:, frame])
            
            # Calculate and display the current time
            current_time = frame * self.dt
            ax.set_title(f'Diffusion over time - {current_time:.2f}s')
            
            return line,
        
        ani = FuncAnimation(fig, update, frames=self.Nt, interval=20, blit=True)
        plt.close(ani._fig)

        return HTML(ani.to_jshtml())
```

```python
# Create an instance of the DiffusionSolver class
solver = DiffusionSolver(T=1, Nt=256, Nx=64, D=0.01, BC='Von-Neumann')

# Solve the diffusion equation
solver.solve_FTCS(BC_values=[-5,0,0,0])

solver.plot_animation()
```
